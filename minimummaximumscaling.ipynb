1.  TargetGuidedOrdinalEncoding

```python
import pandas as pd
```

```python
data_sudhir1=pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
'price':[200,150,300,250,180,320]
})
```

```python
data_sudhir1
```

```python
mean_price=data_sudhir1.groupby('city')['price'].mean()
```

```python
mean_price
```

```python
## calculate the mean price for each city
mean_price=data_sudhir1.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
data_sudhir1['city_encoded']=data_sudhir1['city'].map(mean_price)
```

```python
data_sudhir1
```
